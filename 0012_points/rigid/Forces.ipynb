```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data050 = np.genfromtxt('FD_050.txt',skip_header=1)
data100 = np.genfromtxt('FD_100.txt',skip_header=1)
data200 = np.genfromtxt('FD_200.txt',skip_header=1)
```

```python
Minf = 0.1
Î³ = 1.4
p = 101325.
c = 1.0
```

```python
plt.figure(figsize=(16,20))
angle = np.cumsum(-data050[:,2])

plt.subplot(3,2,1)
plt.plot(data050[:,0],angle)
plt.xlabel('t [s]')
plt.ylabel('Angle [Â°]')
plt.title(r'imposed angle');

plt.subplot(3,2,2)
plt.plot(data050[:,3],data050[:,4])
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title(r'tip position')
plt.axis('equal');

plt.subplot(3,2,3)
plt.plot(data050[:,0],data050[:,7],label='50 punti')
plt.plot(data100[:,0],data100[:,7],label='100 punti')
plt.plot(data200[:,0],data200[:,7],label='200 punti')
plt.legend()
plt.xlabel('t [s]')
plt.ylabel(r'$F_x$ [N]')
plt.title(r'Force x-component');

plt.subplot(3,2,4)
plt.plot(data050[:,0],data050[:,8])
plt.plot(data100[:,0],data100[:,8])
plt.plot(data200[:,0],data200[:,8])
plt.xlabel('t [s]')
plt.ylabel(r'$F_y$ [N]')
plt.title(r'Force y-component');

plt.subplot(3,2,5)
plt.plot(data050[:,0],data050[:,9])
plt.plot(data100[:,0],data100[:,9])
plt.plot(data200[:,0],data200[:,9])
plt.xlabel('t [s]')
plt.ylabel(r'$M_z$ [Nm]')
plt.title(r'Moment z-component');

plt.subplot(3,2,6)
plt.plot(angle*np.pi/180.0,data100[:,8]*2/(p*c*Î³*Minf**2),label=r'$c_l$ computed')
plt.plot(angle*np.pi/180.0,angle*np.pi/180.0*2*np.pi,label=r'$c_l = 2\pi \alpha$')
plt.xlabel('angle [rad]')
plt.legend()
plt.ylabel(r'$c_l$')
plt.title(r'Lift coefficient');
plt.savefig('forces.png')
```
